# Análise de Agrupamentos

Consiste em um conjunto de técnicas que podem ser aplicadas para observar comportamentos/Características semelhantes entre observações. O objetivo nesse tipo de técnica é criar Clusters(grupos) que possuem homogeneidade internamente e heterogenidade entre os grupos.

As técnicas de análise de agrupamentos são consideradas exploratórias, ou de interdependencia, uma vez que sua aplicação não apresentam caráter preditivo para outras observações não presentes no banco de dados. Além disso, a inclusão de novas variáveis, ou observações faz-se necessário que ocorra um rearranj completo das observações nos grupos.

O pesquisador pode optar por elaborar uma análise de agrupamentos quando tiver o objetivo de ordenar e alocar as observações em grupos e, a partir de então, estudar qual a quantidade interessante de clusters formados, ou pode, a priori, definir a quantidade de grupos que deseja formar, embasado por determinado critério, e verificar como se comportam o ordenamento e a alocação das observações naquela quantidade especificada de grupos. Caso o pesquisador deseje utilizar uma técnica para confirmar o estabelecimento de grupos e tornar a análise preditiva, pode-se utilizar Análise Discriminante ou regressão Logística multinomial.

## Medidas de Distância (Dissimilaridade)
É preciso escolher a medida de distância/Semelhança (distância quando métrica e semelhança quando binária) que servirá para considerar se as características são próximas. Para tanto como trabalha-se com distância, é possível que as variáveis estejam em grandezas diferentes, para tanto é preciso padronizar/normalizar as variáveis. As medidas de distâncias mais utilizadas são:

- **Distância Euclidiana:** $d_{pq}=\sqrt{\sum{j=1}^{k}(X_{jp}-X_{jq})^{2}}$. Consiste na medida mais utilizada;
- **Distância Euclidiana Quadrática:** $d_{pq}=\sum{j=1}^{k}(X_{jp}-X_{jq})^{2}$. Frequentemente utilizada quando as variáveis apresentam pequenas dispersão de seus valores.
- **Distância de Minkowski:** $d_{pq}=[\sum{j=1}^{k}(|X_{jp}-X_{jq}|)^{m}]^{\frac{1}{m}}$. É a expressão mais geral, através dela é possível derivar as outras
- **Distância de Manhattan:** $d_{pq}=\sum{j=1}^{k}|ZX_{jp}-ZX_{jq}|$. Consiste na distância absoluta, não leva em consideração a geometria triangular e considera somente as diferenças entre os valores das variáveis.
- **Distância de Chebychev:** $d_{pq}=máx |ZX_{jp}-ZX_{jq}|$. Conhecida como distância máxima, e considera somente a diferença máxima de todas as variáveis.
- **Distância de Canberra:** $d_{pq}= \sum{j=1}^{k}\frac{|ZX_{jp}-ZX_{jq}|}{|ZX_{jp}+ZX_{jq}|}$. É um caso especial, utilizado somente quando as variáveis possuem valores positivos.
- **Coeficiente de correlação:** Está mais para uma medida de similaridade, e pode ser utilizado o coeficiente de correlação em dados faltantes no banco de dados.


##Medidas de Similaridade
Para variáveis binárias utiliza-se medidas de semelhança, que consiste em coeficientes que levam em consideração a similaridade de respostas, sem que necessáriamente esses pares possuam a mesma importância relativa. Primeiro é preciso criar  uma tabela de frequências absolutas de respostas para as variáveis dummy/binária.

   |  | 1 |0| Total|
   |---|---| | |
   |1| a | b| a+b |
   |0| c| d| c+d|
   |Total|a+c|b+d|a+b+c+d|

**As medidas de semelhanças são:**

- **Medida de emparelhamento simples:** Utilizada por Zubin (1938a) e Sokal e Michener (1958). Essa medida, que iguala os pesos das respostas convergentes 1-1 e 0-0, tem sua expressão dada por: $S_{pq}=\frac{a+d}{a+b+c+d}$;

- **Medida de Jaccard:** proposta por Gilbert (1894). Essa medida não leva em conta a frequência do par de resposta 0-0. Contudo, é possível que no banco de dados só tenha frequência para o par 0-0. Nesse caso, a medida de Jaccard é igual a 1. Ela é dada por $S_{pq}=\frac{a}{a+b+c}$;

- **Medida de Dice:** Sugerida e discutida por Czekanowski (1932), Dice (1945) e S0rensen (1948). É similar ao coeficiente de Jaccard, porém dobra o peso sobre a frequência de pares de respostas em convergência do tipo 1-1. é dada por $S_{pq}=\frac{2a}{2a+b+c}$;

- **Medida antiDice:** Proposta por Sokal e Sneath (1963) e Anderberg (1973), a nomenclatura antiDice decorre do fato de que esse coeficiente dobra o peso sobre as frequências de pares de respostas diferentes do tipo 1-1, ou seja, dobra o peso sobre as divergências de respostas. $S_{pq}=\frac{a}{a+2*(b+c)}$;

- **Medida de Russell e Rao:** Proposta por Russell e Rao (1940), a medida privilegia a similaridade de respostas 1-1. Ela é dada por: $S_{pq}=\frac{a}{a+b+c+d}$

- **Medida de Ochiai:** Proposta inicialmente por Driver e Kroeber (1932), e utilizada posteriormente por Ochiai (1957). Esse coeficiente é indefinido quando uma ou ambas as observações estudadas apresentarem os valores de todas as variáveis iguais a 0. Porém, se todos os valores forem iguais a 0, pode-se considerar a medida igual a 1. Sua expressão é dada por: $S_{pq}=\frac{a}{\sqrt{(a+b)*(a+c)}}$

- **Medida de Yule:** Proposta por Yule (1900) e utilizada por Yule e Kendall (1950), essa medida oferece como resposta um coeficiente que varia de -1 a 1. O coeficiente será indefinido se um ou ambos os vetores forem iguais a 0 ou 1. É comum que a medida seja igual a 1 quando b=c=0 (convergencia total de respostas), e igual a -1 quando a=d=0 (divergência total de respostas). É dada por: $S_{pq}=\frac{a*d - b*c}{a*d+b*c}$; 

- **Medida de Rogers e Tanirnoto: Proposta por Rogers e Tanimoto (1960), essa medida dobra o peso das respostas discrepantes 0-1 e 1-0 em relação ao peso das combinações de respostas convergentes do tipo 1-1 e 0-0. É dada por: $S_{pq}=\frac{a+d}{a+d+2*(b+c)}$;

- **Medida de Sneath e Sokal:** Proposta por Sneath e Sokal (1962), dobra o peso das respostas convergentes do tipo 1-1 e 0-0 em relação ao das demais combinações de respostas (1-0 e 0-1). Sua expressão é dada por: $S_{pq}=\frac{2(a+d)}{2(a+d)+b+c}$;

- **Medida de Harnann:**  Hamann (1961) propõe essa medida de semelhança para variáveis binárias com o intuito de que fossem subtraídas as frequências de respostas discrepantes (1-0 e 0-1) do total de respostas convergentes (1-1 e 0-0). Esse coeficiente, que varia de -1 (divergência total de repostas) a 1 (convergência total de respostas). Sua expresão é dada por: $S_{pq}=\frac{(a+d)-(b+c)}{a+b+c+d}$;


## Método Hierárquico Aglomerativo

Nesse método de clusterização, são formados grupamentos cada vez maiores a cada estágio de aglomeração, pela junção de novas observações ou grupos, em função do critério de distância e encadeamento.

O método de encadeamento define qual é método que será utilizado para unir grupamentos com grupamentos ou grupamentos com observações. Existem 3 principais métodos de encadeamento. São eles:

- **Single Linkage/ Nearest Neighbor:** Esse método privilegia a menor distância entre os grupos. É recomendável para casos em que as observações sejam relativamente distantes/Heterogêneas $d_{(MN)W}=mín{(d_{(Mw)};d_{(NW)})}$;

- **Complete Linkage/ Furthest Neighbor:** Esse método privilegia as maiores distâncias entre observações/grupos. É recomendável para casos em que não exista considerável afastamento entre grupos, isto é, quando são muito homogêneos $d_{(MN)W}=máx{(d_{(Mw)};d_{(NW)})}$;

- **Avarege Linkage/ Between Groups:** Esse método leva em consideração a distância média entre todos os pares de observações pertencentes a esse grupo $d_{(MN)W}=\frac{\sum{p=1}^{m+n}\sum{q=1}^{w}d_{pq}}{(m+n)*(w)}$

O dendograma mostra o resultado da clusterização

```python
#Libraries

#----- Data and numerical operations -----#
import numpy as np
import pandas as pd

#----- Graph and Charts -----#
from matplotlib import pyplot as plt
import seaborn as sns

#----- Agglomerative Hierarquical Clustering -----#
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

%matplotlib inline
```

```python
#----- import data -----#
df= pd.read_csv('/cartao_credito.csv', sep=',') #Credit card exemple
df.head(5)
df.columns
```

```python
#----- Normalização -----#
scaler = StandardScaler()
scaler.fit(df.drop(['Sl_No','Customer Key'],axis=1))
scaled_df = scaler.transform(df.drop(['Sl_No','Customer Key'],axis=1))

scaled_df = pd.DataFrame(scaled_df,columns=df.columns[2:]) #remove a ultima  do conjunto
scaled_df.index=df['Customer Key']
scaled_df.head()
```

```python
#------ Dendogram -----#
linkage_data = linkage(scaled_df, method='complete', metric='euclidean') #single| complete| average
dendrogram(linkage_data, above_threshold_color='#bcbddc', orientation='right')
plt.show()
```

```python
#----- Defining Estimators and fitting them to the data -----#
fitted_model = AgglomerativeClustering(n_clusters = 3, linkage = 'complete').fit(scaled_df)
#----- Grouping by clusters -----#
df['Cluster'] = fitted_model.labels_ + 1
group_df = df.groupby('Cluster')
```

```python
#----- Features means by groups -----#
df.groupby('Cluster').mean()
```

```python
#-----  Features variance by groups -----#
df.groupby('Cluster').std()
```
